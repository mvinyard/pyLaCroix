```python
import matplotlib.pyplot as plt, seaborn as sns
from pyLaCroix import pyLaCroix

df = sns.load_dataset('iris')
df.head()
```

```python
palette = pyLaCroix(n_points=1)

for n, i in enumerate(df['species'].unique()):
    
    df_ = df.loc[df['species'] == i]
    plt.plot(range(len(df_['sepal_length'])), df_['sepal_length'], c=palette.kiwi_sandia[n])
    plt.plot(range(len(df_['petal_width'])), df_['petal_width'], c=palette.kiwi_sandia[n+3])
```

```python
palette = pyLaCroix(n_points=5)

for n, i in enumerate(df['petal_width'].unique()):
    
    idx = df.loc[df['petal_width'] == i].index.astype(int)
    plt.scatter(df['sepal_length'][idx], df['sepal_width'][idx], c=palette.apricot[n])
    
```
